```python
import pandas  as pd
import numpy as np
```

```python
# create metadata data

data = pd.DataFrame({})
data["series_uid"] = [i for i in range(2,10)]
data["study_id"] = ["study."+ str(i) for i in range(5001, 5009)]
data["series_desc"] =np.random.choice(["ROI", "Full"], size=len(data))
data["number_of_files"] =np.random.choice([1, 2], size=len(data))
data["file_location"] = ["/files-loc/loc."+ str(i) for i in range(2,10)]

meta_df = pd.DataFrame(data)
meta_df
```

```python
file_paths = []
for loc, files in zip (meta_df["file_location"], meta_df["number_of_files"]):
    for file in range(files):
        file_paths.append(loc + "." + str(file))
len(file_paths)
file_paths
```
